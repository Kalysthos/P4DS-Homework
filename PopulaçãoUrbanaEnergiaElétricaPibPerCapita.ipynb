```python
gdp=pd.read_pickle('New/data.pickle').loc["GDP per capita (current US$)"][[str(num) for num in range(1975,2015)]]
cols = []
for i in range(1975,2015):
    cols.append(str(i))
gdp = gdp[cols].dropna()
gdp_chart=pd.DataFrame()
for i in gdp.index:
        x=pd.DataFrame({'Year':gdp.columns,'Country':[i]*len(gdp.columns),"GDP per capita (current US$)":gdp.T[i]})
        gdp_chart=pd.concat([gdp_chart,x])
gdp_chart=gdp_chart.fillna(0)
gdp_chart
```

```python
import plotly.plotly as py
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 

import pandas as pd
import time
```

```python
List_of_years = list(range(1975,2015))

List_Of_Continents = list(set(Continents.values()))

merged['Year']=merged['Year'].apply(int)

columns = []
# make grid
for year in List_of_years:
    for continent in List_Of_Continents:
        dataset_by_year = merged[merged['Year'] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[dataset_by_year['Continents'] == continent]
        for col_name in dataset_by_year_and_cont:
            # each column name is unique
            column_name = '{year}_{continent}_{header}_gapminder_grid'.format(
                year=year, continent=continent, header=col_name
            )
            a_column = Column(list(dataset_by_year_and_cont[col_name]), column_name)
            columns.append(a_column)

# upload grid
grid = Grid(columns)
url = py.grid_ops.upload(grid, 'GDP.rateplot'+str(time.time()), auto_open=False)
url
```
