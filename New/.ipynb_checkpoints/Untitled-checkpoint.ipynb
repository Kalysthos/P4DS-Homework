```python
import pandas as pd 
from scatter import scatter
import numpy as np
from plots import trend, xsml, trend_years
from collections import defaultdict
data = pd.read_pickle("data.pickle")
```

```python
years = data.columns[::-1]
indices = list(set(data.index.get_level_values(0)))
```

```python
default = (np.nan, np.nan, np.nan, np.nan)
saida = {}
for i in indices:
    year = []
    r2 = []
    coef = []
    amount = []
    curve = []
    for a in years:
        try:
            sca = scatter(data, ['GDP per capita (current US$)',  i], ['2017', a], False)
        except:
            sca = default
        year.append(a)
        r2.append(sca[0])
        coef.append(sca[1])
        amount.append(sca[2])
        curve.append(sca[3])
    a = pd.Series(year, name='year')
    c = pd.Series(r2, name='r2')
    d = pd.Series(coef, name='coef')
    e = pd.Series(amount, name='amount')
    f = pd.Series(curve, name='curve')
    saida[i] = pd.concat([a,c,d,e,f], 1)
```

```python
saida['Interest forgiven (current US$)']
```

```python
total = pd.DataFrame([])
for i in saida.keys():
    acess = saida[i].join(pd.Series([i]*50, name = 'Index'))

    acess.columns = ['Year', 'R2', 'Coefficient', 'Number of countries', 'Type of axis', 'Index']

    acess = acess.set_index(['Index', 'Year']).sort_index()
    
    total = pd.concat([total, acess])
```

```python
final = total.dropna(0).sort_index()
```

```python
final.to_pickle('corrGDP2017.pickle')
```
