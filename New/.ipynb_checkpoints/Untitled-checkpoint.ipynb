```python
import pandas as pd 
from scatter import scatter
import numpy as np
from plots import trend, xsml, trend_years
from collections import defaultdict
data = pd.read_pickle("data.pickle")
```

```python
years = [(i, k) for i in data.columns[::-1] for k in data.columns[::-1]]
indices = list(set(data.index.get_level_values(0)))
```

```python
saida = []
for i in indices:
    for a, b in years:
        try:
            scar = scatter(data, ['GDP per capita (current US$)',  i], [a, b], False)
            if scar[0] >= 0.5 and scar[2] >= 100:
                saida.append(i)
                break
        except:
            pass
```

```python
for i in saida:
    print(i)
```

```python
year1 = []
year2 = []
r2 = []
coef = []
amount = []
curve = []
default = (np.nan, np.nan, np.nan, np.nan)
for i in indices[:1]:
    for a in years:
        for b in years:
            try:
                sca = scatter(data, ['GDP per capita (current US$)',  i], [a, b], False)
            except:
                sca = default
            year1.append(a)
            year2.append(b)
            r2.append(sca[0])
            coef.append(sca[1])
            amount.append(sca[2])
            curve.append(sca[3])
a = pd.Series(year1, name='year1')
b = pd.Series(year2, name='year2')
c = pd.Series(r2, name='r2')
d = pd.Series(coef, name='coef')
e = pd.Series(amount, name='amount')
f = pd.Series(curve, name='curve')
```

```python
pd.concat([a,b,c,d,e,f], 1)
```
